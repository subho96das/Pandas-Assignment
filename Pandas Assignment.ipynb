```python
#6.Rename the column names as
'''Candidate_Name, Marks_Scored, Age_of Candidate, Qualified,
No._of_attempts'''
students ={
   'Candidate Name':['Annie Watson','Katie Blanc','Derek Perry','John Potter','Rose Lee','Matt Aniston','Matthew Cox','Lisa Kudrow','Monica Geller','Rachel Green','Jennifer'],
   'Marks':[9.8,7,6.3,7.2,4.1,5.2,8.5,7.4,6.5,9.5,9.2],'Age':[21,22,20,21,22,23,21,20,19,21,22],'Qualify':['yes', 'yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'yes','yes'],'No. of attempts':[2,1,3,1,2,1,3,4,2,1,2]
}
dfr = pd.DataFrame(students,index= ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j','k'])
dfr=dfr.rename(columns={'Candidate Name':'Candidate_Name','Marks':'Marks_Scored','Age':'Age_of_Candidate','Qualify':'Qualified','No. of attempts':'No._of_attempts'},index={'a':'a','b':'b','c':'c','d':'d','e':'e','f':'f','g':'g','h':'h','i':'i','j':'j','k':'k'})
dfr
```

```python
#7.Split the name and create two new columns name#7.Split the name and create two new columns named - First Name , Last Name

dsp_split=dfr['Candidate_Name'].str.split(' ', expand=False)
print(dsp_split)
dfr['First Name']=dsp_split.str.get(0)
dfr['Last Name']=dsp_split.str.get(1)
dfr
```

```python
#8.Sort the data frame on the basis of Marks
students={'Marks':[9.8,7,6.5,7.2,4.1,5.2,8.5,7.4,6.3,9.5,9.2]}
print("*************sorting on the basis of marks in ascending order***************")
u_df=pd.DataFrame(students,index=['a','b','c','d','e','f','g','h','i','j','k'])
sor_df=u_df.sort_values(by='Marks')
print(sor_df)
```

```python
#10.Group the data on the basis of Age and print the mean, max, sum of the marks of that group
score={'Age_of_Candidate':[19,20,20,21,21,21,21,22,22,22,23],'Marks_Scored':[6.3,6.5,7.4,7.2,8.5,9.5,9.8,4.1,7.0,9.2,5.2]}
dfr = pd.DataFrame.from_dict(score)
gr=dfr.groupby('Age_of_Candidate')
print(gr['Marks_Scored'].agg([np.mean,np.max,np.sum]))
```
